8. Выведите минимальное и максимальное представимое значение для каждого скалярного типа numpy

````python
typechars = ['S1', '?', 'B', 'D', 'G', 'F', 'I', 'H', 'L', 'O', 'Q',
             'S', 'U', 'V', 'b', 'd', 'g', 'f', 'i', 'h', 'l', 'q']
for typechar in typechars:
    if np.issubdtype(np.dtype(typechar), np.integer):
        print(typechar, ' : ', np.typename(typechar))
        print(np.iinfo(np.dtype(typechar)))

    elif np.issubdtype(np.dtype(typechar), np.floating):
        print(typechar, ' : ', np.typename(typechar))
        print( np.finfo(np.dtype(typechar)))
````

3. Правда ли, что люди с высоким уровнем стресса имеют более высокое кровяное давление, чем те, кто не подвержен стрессу

````python
print("Lets's consider, that the high level of stress is value >50, and values are <= 50 is the indicator of low stress level.Then:")
print(datset[datset['Stress Levels']>= 50])
print()
print("We got the emty dataframe, so we haven't values > 50/ Lets's check max and min values for this column")
print(datset['Stress Levels'].describe())

print("Let's slip elements for the the midle vaue:")
mid_val =datset['Stress Levels'].min() + (datset['Stress Levels'].max() - datset['Stress Levels'].min())/2
print("Middle val= ", mid_val)

datset["Systolic_pressure"] = (datset['Blood Pressure (s/d)'].str.split('/').str[0].astype(int))
datset["Diastolic_pressure"] = (datset['Blood Pressure (s/d)'].str.split('/').str[1].astype(int))

print()
print("High stress:")
mid_high_press_s = datset[datset['Stress Levels']> mid_val]['Systolic_pressure'].mean()
mid_high_press_d = datset[datset['Stress Levels']> mid_val]['Diastolic_pressure'].mean()
print("Systolic = ", mid_high_press_s, "Diastolic = ", mid_high_press_d)
print("Low stress:")
mid_low_press_s = datset[datset['Stress Levels']<= mid_val]['Systolic_pressure'].mean()
mid_low_press_d = datset[datset['Stress Levels']<= mid_val]['Diastolic_pressure'].mean()
print("Systolic = ", mid_low_press_s,  "Diastolic = ", mid_low_press_d )
print("Diff = ", mid_high_press_s-mid_low_press_s, " | ", mid_high_press_d - mid_low_press_d)
print()
print("We got a little different of values, but at the both case? people with high stress level has a large values ")
````
